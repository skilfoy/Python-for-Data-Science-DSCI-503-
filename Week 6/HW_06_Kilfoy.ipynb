```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('iris.txt', sep='\t')
ames = pd.read_csv('ames_housing.txt', sep='\t')
titanic = pd.read_csv('titanic.txt', sep='\t')
batting = pd.read_csv('batting.csv', sep=',')
```

## Problem 1: Iris Dataset

```python
iris.head(8)
```

```python
iris.describe()
```

```python
iris_means = iris.groupby('species').mean()
iris_means
```

```python
iris_colors = ['cornflowerBlue',
               'slateBlue',
               'darkOrchid']
iris_columns = ['Sepal Length',
                'Sepal Width',
                'Petal Length',
                'Petal Width']
```

```python
fig, axs = plt.subplots(2, 2, figsize=[8,6])
for i, ax in enumerate(axs.flatten()):
    col_data = iris_means.iloc[:, i]
    ax.bar(iris_means.index,
           col_data,
           color=iris_colors,
           edgecolor='black')
    ax.set_title('Mean ' + iris_columns[i])
    ax.set_ylabel(iris_columns[i])
plt.tight_layout()
plt.show()
```

## Problem 7: Batting Data  Batting Averages

```python
batting_avg = batting.groupby(['teamID', 'yearID'])[['H', 'AB']].sum()
batting_avg['BA'] = batting_avg['H'] / batting_avg['AB']
batting_avg = batting_avg.sort_values('BA', ascending=False)
batting_avg.head(10)
```

## Problem 8: Batting Data  Cardinals vs. Cubs

```python
stl_batting = batting.loc[(batting['teamID'] == 'SLN') & (batting['yearID'] >= 1900)].loc[:, ['yearID', 'H', 'AB', 'HR']]
stl_batting = stl_batting.groupby('yearID').sum()
stl_batting['BA'] = stl_batting['H']/stl_batting['AB']
chi_batting = batting.loc[(batting['teamID'] == 'CHN') & (batting['yearID'] >= 1900)].loc[:, ['yearID', 'H', 'AB', 'HR']]
chi_batting = chi_batting.groupby('yearID').sum()
chi_batting['BA'] = chi_batting['H']/chi_batting['AB']

plt.figure(figsize=[12, 4])
plt.subplot(1,2,1)
plt.plot(stl_batting.index,
         stl_batting['BA'],
         color='crimson',
         label='Cardinals')
plt.plot(chi_batting.index,
         chi_batting['BA'],
         color='royalblue',
         label='Cubs')
plt.xlabel('Year')
plt.ylabel('Batting Average')
plt.title('Batting Average By Year')
plt.legend()

plt.subplot(1,2,2)
plt.plot(stl_batting.index,
         stl_batting['HR'],
         color='crimson',
         label='Cardinals')
plt.plot(chi_batting.index,
         chi_batting['HR'],
         color='royalblue',
         label='Cubs')
plt.xlabel('Year')
plt.ylabel('Home Runs')
plt.title('Home Runs by Year')
plt.legend()

plt.show()
```

```python
prop_batting = round(np.mean(stl_batting['BA'] > chi_batting['BA']), 4)
print('Proportion of years since 1900 in which the\nCardinals had a higher batting average than the Cubs:')
prop_batting
```

```python
prop_hr = round(np.mean(stl_batting['HR'] > chi_batting['HR']), 4)
print('Proportion of years since 1900 in which the\nCardinals had more home runs than the Cubs:')
prop_hr
```
